Several medical features contain zero values that are physiologically impossible. These zeros are likely placeholders for missing measurements and should be treated as missing values before training the model.

```python
# Replace null values with median

import numpy as np

cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in cols:
    df[col] = df[col].fillna(df[col].median())
```

# Define Features and Target

```python
x = df.drop("Outcome", axis = 1 )
y = df["Outcome"]
```

```python
x.shape
y.shape
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,random_state = 42, stratify  = y)
```

We have used Stratify function as It ensures that both the training and test sets maintain approximately the same proportion of diabetic and non-diabetic patients as the original dataset.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression(class_weight = "balanced")

log.fit(x_train_scaled, y_train)
y_pred = log.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("classification report:", classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_prob = log.predict_proba(x_test_scaled)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_prob)

auc = roc_auc_score(y_test, y_prob)

print(auc)
```

ROC-AUC Score: The Logistic Regression model achieved an AUC score of 0.82, indicating a good ability to distinguish between diabetic and non-diabetic patients. An AUC of 0.82 means that if one diabetic patient and one non-diabetic patient are selected at random, the model has an 82% probability of assigning a higher predicted probability to the diabetic patient.

```python
plt.figure(figsize=(12,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')


plt.title("ROC Curve")
plt.xlabel ("FPR")
plt.ylabel("TPR")
plt.legend()


```

### ROC Curve Observation

The ROC curve lies well above the diagonal reference line, indicating that the Logistic Regression model performs significantly better than random classification. The model achieved an **AUC score of 0.82**, which indicates a good ability to distinguish between diabetic and non-diabetic patients. Overall, the ROC analysis suggests that the model is effective in ranking patients based on their probability of having diabetes.


```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

skf = StratifiedKFold( n_splits = 5, shuffle = True, random_state  =42 )

cvs = cross_val_score( log, x, y, cv = skf, scoring = 'accuracy')

print("Accuracy of each fold:", cvs )
print("Average Accuracy:", cvs.mean())
print("Standard Deviation:", cvs.std())
```

The Logistic Regression model achieved an average cross-validation accuracy of 74.73% with a standard deviation of 4.91% using Stratified 5-Fold Cross-Validation. Since the cross-validation accuracy is close to the test accuracy (73.38%), the model demonstrates good stability and generalization across different subsets of the dataset.

```python
# HYPER PARAMETER TUNING

from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs'],
    'class_weight': [None, 'balanced'],
    'max_iter': [1000]
}

grid = GridSearchCV(
    estimator=LogisticRegression(),
    param_grid=param_grid,
    scoring='recall',
    cv=5,
    n_jobs=-1
)

grid.fit(x_train_scaled, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Cross Validation Recall:", grid.best_score_)
```

```python
best_model = grid.best_estimator_
```

```python
y_pred = best_model.predict(x_test_scaled)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall   :", recall_score(y_test, y_pred))
print("F1 Score :", f1_score(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report")
print(classification_report(y_test, y_pred))
```

Hyperparameter tuning was performed using GridSearchCV to identify the optimal Logistic Regression parameters. Compared to the balanced Logistic Regression model, the tuned model achieved a recall of 72.2%, improving its ability to correctly identify diabetic patients. Although the overall accuracy decreased slightly from 73.4% to 72.7%, the number of False Negatives reduced from 16 to 15, which is beneficial for a medical screening application where missing a diabetic patient is more critical than generating additional False Positives. Therefore, the tuned model was selected as the preferred model because it better aligns with the project's objective of minimizing Type II errors.

# Final Observation


### Final Model Selection

Three Logistic Regression models were evaluated: the baseline model, the model with `class_weight='balanced'`, and the hyperparameter-tuned model. The models were compared using Accuracy, Precision, Recall, F1-Score, ROC-AUC, and Cross-Validation.

For this diabetes prediction problem, the main objective is to **reduce False Negatives (Type II errors)**. A False Negative means that a patient who actually has diabetes is predicted as not having diabetes. This could delay treatment and lead to serious health complications. Although False Positives may cause unnecessary follow-up tests, they are generally less harmful than missing a diabetic patient.

Therefore, **Recall** was considered the most important evaluation metric. The hyperparameter-tuned Logistic Regression model achieved the highest Recall (72.2%) and reduced the number of False Negatives compared to the other models. Although its accuracy was slightly lower than the balanced Logistic Regression model, it better satisfied the business objective.

Hence, the **hyperparameter-tuned Logistic Regression model** was selected as the final model for predicting diabetes on unseen data because it best aligns with the goal of minimizing Type II errors.
